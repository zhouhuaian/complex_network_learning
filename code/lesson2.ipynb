```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 绘制ER随即网络
n = 10000  # 节点数目
p = 0.001  # 链接生成概率
er_g = nx.erdos_renyi_graph(n, p)
```

```python
d = dict(er_g.degree)
print("平均度: ", sum(d.values()) / n)  # <k> = (n-1)p 
```

```python
# 绘制ER随机网络度分布
x = [i for i in range(max(d.values()) + 1)]
y = [i / n for i in nx.degree_histogram(er_g)]

plt.plot(x, y, "bo-")
plt.xlabel("$k$")
plt.ylabel("$p_k$")
plt.show()
```

```python
# 绘制BA无标度网络
m = 3
ba_g = nx.barabasi_albert_graph(n, m)
```

```python
d = dict(nx.degree(ba_g))
print("平均度为: ", sum(d.values()) / n)
```

```python
# 绘制BA无标度网络度分布
x = [i for i in range(max(d.values()) + 1)]
y = [i / n for i in nx.degree_histogram(ba_g)]

plt.plot(x, y, "go-")
plt.xlabel("$k$")
plt.ylabel("$p_k$")
```

```python
# 在log-log坐标轴下绘制
plt.plot(x, y, "yo-")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("$k$")
plt.ylabel("$p_k$")
```

```python
# 因为log0无意义，所以删除掉p_k=0的点
new_x, new_y = [], []
for i in range(len(x)):
    if y[i] != 0:
        new_x.append(x[i])
        new_y.append(y[i])

plt.plot(new_x, new_y, "go-")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("$k$")
plt.ylabel("$p_k$")
```

```python
print("网络的直径为：", nx.diameter(ba_g))  # 计算网络直径
print("网络平均距离：", nx.average_shortest_path_length(ba_g))  # 网络的平均距离
```

```python
print(nx.shortest_path_length(ba_g, 1, 5))  # 计算节点间距离
print(nx.efficiency(ba_g, 1, 5))  # 计算节点间效率（=1/距离）
```

```python
print(nx.global_efficiency(ba_g))  # 图的全局效率（所有节点对间效率平均值）
print(nx.local_efficiency(ba_g))  # 图的局部效率（所有节点局部效率的平均值）
```
