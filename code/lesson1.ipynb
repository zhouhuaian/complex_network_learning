```python
import networkx as nx
```

```python
edge_list = [(1, 2), (1, 3), (2, 3), (2, 4)]
G = nx.Graph(edge_list)  # 生成无向图
```

```python
nx.draw(G, node_size=500, with_labels=True)  # 绘图
```

```python
ad_mat = nx.adjacency_matrix(G)  # 邻接矩阵
print(ad_mat)
print(type(ad_mat))
```

```python
a = ad_mat.todense()  # 转换为矩阵形式
print(a)
```

```python
import numpy as np
```

```python
A = np.array([[0, 1, 1], [1, 0, 1], [1, 1, 0]])
G = nx.from_numpy_matrix(A)  # 由邻接矩阵生成图
nx.draw(G, node_size=500, with_labels=True)
```

```python
G = nx.Graph()
G.add_weighted_edges_from([(1, 2, 0.75), (1, 3, 1.3), (1, 4, 0.3)])  # 生成加权图
nx.draw(G, with_labels=True)
```

```python
edge_list = [[1, 2], [2, 3], [2, 4], [3, 4]]
G = nx.Graph(edge_list)
nx.draw(G, with_labels=True)
```

```python
G.degree  # 节点的度
```

```python
d = dict(G.degree)
print(d)
```

```python
print("平均度: ", sum(d.values()) / len(G.nodes))
```

```python
nx.degree_histogram(G)  # 度的频数分布
```

```python
import matplotlib.pyplot as plt
```

```python
x = [i for i in range(max(d.values()) + 1)]
y = [i / len(G.nodes) for i in nx.degree_histogram(G)]
print(x)
print(y)
```

```python
plt.bar(x, y, width=0.5, color='red')  # 绘制度分布的直方图
plt.ylabel("$p_k$")
plt.xlabel("$k$")
plt.xlim(0, 4)
plt.show()
```

```python
G = nx.Graph()
G.add_nodes_from(range(1, 6))
G.add_edges_from([(1, 2), (2, 3), (2, 5), (3, 4), (4, 5)])
nx.draw(G, with_labels=True)
```

```python
print(nx.shortest_path(G, source=1, target=4))  # 最短路径
print(nx.shortest_path_length(G, source=2, target=4))  # 最短路径长度
```

```python
list(nx.all_shortest_paths(G, source=1, target=4))  # 所有的最短路径
```

```python
nx.average_shortest_path_length(G)
```

```python
Ga = nx.Graph()
Ga.add_edges_from([(1, 2), (1, 3), (2, 3), (4, 7), (5, 6), (5, 7), (6, 7)])
nx.draw(Ga, node_size=500, with_labels=True)
```

```python
edge_list = ((1, 2), (1, 3), (2, 3), (2, 4), (4, 7), (5, 6), (5, 7), (6, 7))
Gb = nx.Graph(edge_list)
nx.draw(Gb, node_size=500, with_labels=True)
```

```python
print(nx.is_connected(Ga))  # 判断图是否为连通图
print(nx.is_connected(Gb))
```

```python
edge_list = [(1, 2), (1, 3), (1, 4), (1, 5), (2, 3), (2, 4), (2, 5), (3, 4), (3, 5), (4, 5)]
G = nx.Graph(edge_list)
nx.draw(G, node_size=500, with_labels=True)
```

```python
print(nx.clustering(G, 1))  # 计算节点集聚系数
```

```python
G.remove_node(2)
nx.draw(G, with_labels=True)
```

```python
G.remove_edge(1, 3)
nx.draw(G, with_labels=True)
```

```python
print(nx.clustering(G))
```

```python
print(nx.average_clustering(G))  # 图的平均集聚系数
```

```python
print(nx.transitivity(G))  # 计算全局集聚系数（即图的传递性）
```
