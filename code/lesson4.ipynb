```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 自定义k近邻规则网络
def regular_net(n, k):
    g = nx.Graph()
    nodes = list(range(n))  # 0 ~ n-1
    # 顺时针方向，每个节点和相邻k/2个节点相连
    for i in range(1, k // 2 + 1):
        targets = nodes[i: ] + nodes[0 : i]
        g.add_edges_from(zip(nodes, targets))
    return g
```

```python
n = 20
k = 4
rg = regular_net(n, k)  # 4近邻规则图
pos = nx.circular_layout(rg)
nx.draw(rg, pos, node_size=100, node_color="orange", edge_color="gray", with_labels=False)
plt.title("k-regular")
```

```python
# 绘制规则网络
p1 = 0
rg = nx.watts_strogatz_graph(n, k, p1)  # 重连概率为0，生成k近邻规则网络
pos1 = nx.circular_layout(rg)
nx.draw(rg, pos1, node_size=100, node_color="green", edge_color="gray", with_labels=False)
plt.title("k-regular")
```

```python
# 绘制WS小世界网络
p2 = 0.2
ws_g = nx.watts_strogatz_graph(n, k, p2)
pos2 = nx.circular_layout(ws_g)
nx.draw(ws_g, pos2, node_size=100, node_color="pink", edge_color="gray", with_labels=False)
plt.title("WS-small-world")
```

```python
# 绘制随机网络
p3 = 1
random_g = nx.watts_strogatz_graph(n, k, p3)  # 重连概率为1，生成随机网络
pos3 = nx.circular_layout(random_g)
nx.draw(random_g, pos3, node_size=100, node_color="purple", edge_color="gray", with_labels=False)
plt.title("random")
```

```python
# 绘制k近邻规则网络
plt.figure(figsize=(15, 5))
p1 = 0
rg = nx.newman_watts_strogatz_graph(n, k, p1)  # 添边概率为0，生成k近邻规则网络
pos1 = nx.circular_layout(rg)
plt.subplot(131)
nx.draw(rg, pos1, node_size=100, node_color="red", edge_color="gray", with_labels=False)
plt.title("k-regular")

# 绘制NW小世界网络
p2 = 0.3
nw_g = nx.newman_watts_strogatz_graph(n, k, p2)  
pos2 = nx.circular_layout(nw_g)
plt.subplot(132)
nx.draw(nw_g, pos2, node_size=100, node_color="yellow", edge_color="gray", with_labels=False)
plt.title("NW-small-world")

# 绘制完全图
p3 = 1
complete = nx.newman_watts_strogatz_graph(n, k, p3)  # 添边概率为1，生成完全图
pos3 = nx.circular_layout(complete)
plt.subplot(133)
nx.draw(complete, pos3, node_size=100, node_color="blue", edge_color="gray", with_labels=False)
plt.title("complete")
```
